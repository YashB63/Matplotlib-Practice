```python
# Ab baat karte hai transparency ki

import matplotlib.pyplot as plt
day = [1,2,3,4,5,6,7]
number = [2,3,1,4,5,3,6]
plt.scatter(day,number)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)
plt.show()
```

```python
colors = ["r","y","g","b","m","cyan","orange"]
sizes = [100,200,400,500,200,100,700]
plt.scatter(day,number,c=colors,s=sizes)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)
```

```python
# alpha parameter use karege
colors = ["r","y","g","b","m","cyan","orange"]
sizes = [100,200,400,500,200,100,700]
plt.scatter(day,number,c=colors,s=sizes,alpha=0.2)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)
```

```python
colors = ["r","y","g","b","m","cyan","orange"]
sizes = [100,200,400,500,200,100,700]
plt.scatter(day,number,c=colors,s=sizes,alpha=0.4)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)
```

```python
# plot mey dots ke jagah koi aur design chahiye to marker parameter use karege
colors = ["r","y","g","b","m","cyan","orange"]
sizes = [100,200,400,500,200,100,700]
plt.scatter(day,number,c=colors,s=sizes,marker="*")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)
```

```python
# design ki border ko change karna hai to edge color parameter use karege
colors = ["r","y","g","b","m","cyan","orange"]
sizes = [1000,20000,4000,5000,2000,1000,7000] #edge color dikhane ke liye size badhayi hai
plt.scatter(day,number,c=colors,s=sizes,marker="*",edgecolor="g")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)
```

```python
# edge ko mota karna ho to linewidth parameter use karege
colors = ["r","y","g","b","m","cyan","orange"]
sizes = [100,200,400,500,200,100,700]
plt.scatter(day,number,c=colors,s=sizes,marker="*",edgecolor="g",linewidth=4)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)
```

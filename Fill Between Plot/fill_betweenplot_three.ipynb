```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5]) # convert the list into numpy array
y = np.array([1,2,3,4,5])
```

```python
# Shaded region ko limit karna ho to
plt.plot(x,y,color="black")
plt.fill_between(x,y,color="m",where=(x>=2)&(x<=4)) # ab yaha array ka concept use ho rha hai(where ke andar ke parameters) 
plt.title("Python")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```

```python
# color light karna ho to use alpha parameter
plt.plot(x,y,color="black")
plt.fill_between(x,y,color="m",where=(x>=2)&(x<=4),alpha=0.5) # ab yaha array ka concept use ho rha hai(where ke andar ke parameters) 
plt.title("Python")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```

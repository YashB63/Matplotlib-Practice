```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
```

```python
plt.plot(x,y)
plt.show()
```

```python
# color change karna ho to use color parameter
plt.plot(x,y,color='r')
plt.show()
```

```python
# title lagana ho to use title parameter
plt.plot(x,y,color="r")
plt.title("Python")
plt.show()
```

```python
# label lagana ho to use label parameter
plt.plot(x,y,color="r")
plt.title("Python")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
```

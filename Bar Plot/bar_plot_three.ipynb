```python
import matplotlib.pyplot as plt
x = ["python","c","c++", "java"]
y = [85,70,60,82]
plt.xlabel("language",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.title("bar_plot_three",fontsize=15)
plt.bar(x,y,width=0.4,color="m")
plt.show()
```

```python
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color=c,edgecolor="r",linewidth=5)
```

```python
# Agar bar ke colors ko light karna ho to alpha parameter use karege
plt.bar(x,y,width=0.4,color=c,edgecolor="r",linewidth=5,alpha=0.5)
```

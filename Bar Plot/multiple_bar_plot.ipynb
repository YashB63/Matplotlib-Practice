```python
import matplotlib.pyplot as plt
x = ["python","c","c++", "java"]
y = [85,70,60,82]
plt.xlabel("language",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.title("multiple_bar_plot",fontsize=15)
plt.bar(x,y,width=0.4,color="m")
plt.show()
```

```python
#c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color="r",edgecolor="r",label="popularity")
plt.legend()
```

```python
# ek aur parameter introduce karte hai
z = [20,30,40,50]
plt.bar(x,y,width=0.4,color="r",label="popularity")
plt.bar(x,z,width=0.4,color="y",label="popularity1")
plt.legend()
```

```python
import matplotlib.pyplot as plt
x = ["python","c","c++", "java"]
y = [85,70,60,82]
plt.xlabel("language",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.title("multiple_bar_three",fontsize=15)
plt.bar(x,y,width=0.4,color="m")
plt.show()
```

```python
z = [20,30,40,50]
width = 0.2
plt.bar(x,y,color="r",label="popularity")
plt.bar(x,z,color="y",label="popularity1")
plt.legend()
```

```python
import numpy as np
p = np.arange(len(x))
plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p,z,width,color="y",label="popularity1")
plt.legend()
```

```python
# pehle x axis mey python c++ wagera likha hua aa rha tha ab 0, 1, 2 aise numbers aa rhe hai
p1 = [j+width for j in p] # ye step ke baad 2 bar plot side by side ban jaayege
plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p1,z,width,color="y",label="popularity1")
plt.legend()
```

```python
# abhi bhi naam nazar nahi aa rhe hai x axis mey python wagera
p1 = [j+width for j in p] # ye step ke baad 2 bar plot side by side ban jaayege
plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p1,z,width,color="y",label="popularity1")
plt.xticks(p+width,x) # iski madad se naam nazar aayege
plt.legend()
```

```python
# Ab yaha ek aur problem hai. Abhi naam to nazar aa rhe hai. uske liye width/2 karege
plt.xlabel("language",fontsize=15)
plt.ylabel("number",fontsize=15)
p1 = [j+width for j in p] # ye step ke baad 2 bar plot side by side ban jaayege
plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p1,z,width,color="y",label="popularity1")
plt.xticks(p+width/2,x) # iski madad se naam nazar aayege
plt.legend()
```

```python
# ab agar x axis ke naam ko ghumana ho to
plt.xlabel("language",fontsize=15)
plt.ylabel("number",fontsize=15)
p1 = [j+width for j in p] # ye step ke baad 2 bar plot side by side ban jaayege
plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p1,z,width,color="y",label="popularity1")
plt.xticks(p+width/2,x,rotation=15) # iski madad se naam nazar aayege
plt.legend()
```

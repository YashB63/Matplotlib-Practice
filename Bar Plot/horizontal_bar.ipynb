```python
import matplotlib.pyplot as plt
x = ["python","c","c++", "java"]
y = [85,70,60,82]
z = [20,30,40,50]

plt.xlabel("language",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.title("horizontal_bar",fontsize=15)

plt.barh(x,y,color="r",label="popularity")
plt.barh(x,z,color="y",label="popularity1")
plt.legend()
plt.show()
```

```python
import numpy as np
p = np.arange(len(x))

plt.xlabel("language",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.title("horizontal_bar",fontsize=15)

width=0.4
p1 = [j+width for j in p]
plt.barh(p,y,color="r",label="popularity")
plt.barh(p1,z,color="y",label="popularity1")
plt.legend()
plt.show()
```

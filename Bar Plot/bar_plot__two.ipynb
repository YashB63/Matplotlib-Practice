```python
import matplotlib.pyplot as plt
x = ["python","c","c++", "java"]
y = [85,70,60,82]
plt.xlabel("language",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.title("bar_plot_two",fontsize=15)
plt.bar(x,y,width=0.4)
plt.show()
```

```python
#Highlighting the bar plot with red color
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color=c,edgecolor="r")
```

```python
#highlight color ko mota karne ke liye linewidth use karege
plt.bar(x,y,width=0.4,color=c,edgecolor="r",linewidth=5)
```

```python
# agar -- waala highlight chahiye to linestyle use karege
plt.bar(x,y,width=0.4,color=c,edgecolor="r",linewidth="5",linestyle=":")
```

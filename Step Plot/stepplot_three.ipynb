```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [5,3,1,7,2]
```

```python
plt.step(x,y,color="r",marker="o",ms=20,mfc="cyan")
plt.grid()
plt.title("Python")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```

```python
# Label Parameter
plt.step(x,y,color="r",marker="o",ms=20,mfc="cyan",label="Python")
plt.grid()
plt.title("Python")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.legend() # label dekhne ke liye legend function use karna pdhta hai
plt.show()
```

```python
# Ab agar aap chahte hai ki legend ki position change ki jaaye to loc parameter use kare
# marker ka color change karna ho to mfc parameter use karo
plt.step(x,y,color="r",marker="o",ms=20,mfc="cyan",label="Python")
plt.grid()
plt.title("Python")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.legend(loc=2) # loc mey values daalo and legend ki position change karo 
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
```

```python
plt.step(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,5,7,1,3]
```

```python
plt.step(x,y)
plt.show()
```

```python
# grid daalna ho to chart mey
plt.step(x,y)
plt.grid()
plt.show()
```

```python
# Title daalna ho to
plt.step(x,y)
plt.title("Python")
plt.grid()
plt.show()
```

```python
# Label daalna ho to
plt.step(x,y)
plt.title("Python")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [5,3,1,7,2]
```

```python
plt.step(x,y)
plt.grid()
plt.title("Python")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```

```python
# marker parameter
plt.step(x,y,marker="o")
plt.grid()
plt.title("Python")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```

```python
# Plot ka colour change karna ho to use colour parameter
plt.step(x,y,color="r",marker="o")
plt.grid()
plt.title("Python")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```

```python
# Marker ka size badhana ho to use ms parameter
plt.step(x,y,color="r",marker="o",ms=20)
plt.grid()
plt.title("Python")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```

```python
# marker ka color change karna ho to mfc parameter use karo
plt.step(x,y,color="r",marker="o",ms=20,mfc="m")
plt.grid()
plt.title("Python")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```

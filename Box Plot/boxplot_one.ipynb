```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60,70]
```

```python
plt.boxplot(x)
plt.show()
```

```python
# Hourglass shape banana ho to
plt.boxplot(x,notch=True)
plt.show()
```

```python
# Horizontal boxplot badhana ho to
plt.boxplot(x,vert=False)
plt.show()
```

```python
# width parameter agar box ki width badhani ho to
plt.boxplot(x,widths=0.3)
plt.show()
```

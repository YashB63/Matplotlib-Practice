```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60,70]
# box ko label dena ho to use labels parameter and enter the value inside list
plt.boxplot(x,labels=["Python"])
plt.show()
```

```python
# box mey color fill karna ho to use patch_artist parameter and set it to true
plt.boxplot(x,labels=["Python"], patch_artist=True)
plt.show()
```

```python
# agar aap mean ka ek marker dekhna chahte hai to use showmeans parameter and set it to True
plt.boxplot(x,labels=["Python"],showmeans=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=["Python"],showmeans=True)
plt.show() # o symbol represents outlier
```

```python
# ab outlier aur line ko saath mey join karna ho to we use whis parameter
plt.boxplot(x,labels=["Python"],showmeans=True,whis=2)
plt.show() # ab aap dekh skte ho ki outlier line ke saath connect ho gaya hai
```

```python
# ab outlier ka colour aur shape change karna ho to we use sym parameter
plt.boxplot(x,labels=["Python"],showmeans=True,sym="m*") #colour magenta and shape star
plt.show() 
```

```python
# outlier ko hide karna ho to
plt.boxplot(x,labels=["Python"],showmeans=True,sym=" ") 
plt.show() # outlier nazar nahi aayega
```

```python
# ab box ka color change karna ho to use boxprop parameter and enter values inside dictionary
plt.boxplot(x,labels=["Python"],showmeans=True,sym="m*",boxprops=dict(color="r"))
plt.show() 
```

```python
# ab box ke upar niche jpo line caps hai(horizontal lines) unka color change karna ho to
# use capprops parameter and enter the values inside dictionary
plt.boxplot(x,labels=["Python"],showmeans=True,sym="m*",boxprops=dict(color="r"),capprops=dict(color="r")) 
plt.show()
```

```python
# Ab agar aap chahte hai ki whisker(vertical lines) ka bhi color change ho to use whiskerprops parameter. Values dictionary ke andar daalni hai
plt.boxplot(x,labels=["Python"],showmeans=True,sym="m*",boxprops=dict(color="r"),capprops=dict(color="r"),whiskerprops=dict(color="r")) 
plt.show() # purey box ka color change ho chuka hai.
```

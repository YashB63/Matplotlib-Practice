```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xticks(x)
plt.yticks(y)
plt.show() 
```

```python
# kisi axis par custom labels daalne ho to use labels parameter inside xticks or yticks
x = [1,2,3,4]
y = [3,1,5,2]
plt.plot(x,y)
plt.xticks(x,labels=["python","c","c++","java"])
plt.yticks(y)
plt.show()
```

```python
# agar aap kisi axis ki limit badhana chahate ho to 
x = [1,2,3,4]
y = [3,1,5,2]
plt.plot(x,y)
plt.xlim(0,10) # ab aap dekhoge ki x axis ki limit badh gyi hai
plt.show()
```

```python
# y ki limit change karna ho to use ylim
x = [1,2,3,4]
y = [3,1,5,2]
plt.plot(x,y)
plt.ylim(0,20) # ab aap dekhoge ki y axis ki limit badh gyi hai
plt.show()
```

```python
# Axis ke andar agar aap x aur y axis ke limit change karna chahte hai to use a list inside axis and add parameters like wise
x = [1,2,3,4]
y = [3,1,5,2]
plt.plot(x,y)
plt.axis([0,10,0,7]) # (x_axis_limit_start,x_axis_limit_end,y_axis_limit_start,y_axis_limit__end)
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]
```

```python
plt.plot(x,y)
plt.show()
```

```python
# Agar aap chahte hai ki jo values x mey aapne add ki hai vahi x axis mey dikhe to uske liye use xticks parameter
plt.plot(x,y)
plt.xticks(x)
plt.show()
```

```python
# Similarly agar aap chahte hai ki jo values aapne y mey add ki hai vahi y axis mey dikhe to use yticks parameter
plt.plot(x,y)
plt.yticks(y)
plt.show() 
```

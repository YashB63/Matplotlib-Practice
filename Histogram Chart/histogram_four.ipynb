```python
import matplotlib.pyplot as plt
```

```python
num = [17,17,23,58,43,12,20,11,38,34 ,54, 32, 55, 30, 43, 17, 21, 33, 55, 20, 20, 29, 28, 33,
 49, 39, 23, 50, 34, 45, 58, 14, 36, 14, 46, 10, 18, 20, 14, 52, 24, 46, 57, 43, 42, 12, 21, 14,
 16, 38]
plt.hist(num)
plt.show()
```

```python
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num)
plt.show()
```

```python
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num, color="r", edgecolor="black")
plt.show()
```

```python
# bottom parameter
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num, color="r", edgecolor="black", bottom=20)
plt.show()
```

```python
# align parameter
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num, color="r", edgecolor="black", align="left")
plt.show()
```

```python
#  changing type of histogram
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num, color="r", edgecolor="black", histtype="step")
plt.show()
```

```python
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num, color="r", edgecolor="black",histtype="stepfilled")
plt.show()
```

```python
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num, color="r", edgecolor="black", histtype="barstacked")
plt.show()
```

```python
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num, color="r", edgecolor="black", histtype="bar")
plt.show()
```

```python
# Changing orientation
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num, color="r", edgecolor="black", orientation = "horizontal")
plt.show()
```

```python
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num, color="r", edgecolor="black", orientation="vertical") # by default vertical horta hi hai
plt.show()
```

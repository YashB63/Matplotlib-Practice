```python
import matplotlib.pyplot as plt
```

```python
num = [17,17,23,58,43,12,20,11,38,34 ,54, 32, 55, 30, 43, 17, 21, 33, 55, 20, 20, 29, 28, 33,
 49, 39, 23, 50, 34, 45, 58, 14, 36, 14, 46, 10, 18, 20, 14, 52, 24, 46, 57, 43, 42, 12, 21, 14,
 16, 38]
plt.hist(num)
plt.show()
```

```python
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num)
plt.show()
```

```python
# ab iska colour change karte hai
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num,color="r") #yaha changes karne se color change hoga histogram ka
plt.show()
```

```python
# ab yaha ek aur pareshani hai. saare bars chipke hue hai kuch smjh ni aa rha. 
# ab agar aap bar seperately dekhna chahte hai to use edge color parameter
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num, color="r", edgecolor="black")
plt.show()
```

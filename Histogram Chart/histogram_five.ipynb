```python
import matplotlib.pyplot as plt
```

```python
num = [17,17,23,58,43,12,20,11,38,34 ,54, 32, 55, 30, 43, 17, 21, 33, 55, 20, 20, 29, 28, 33,
 49, 39, 23, 50, 34, 45, 58, 14, 36, 14, 46, 10, 18, 20, 14, 52, 24, 46, 57, 43, 42, 12, 21, 14,
 16, 38]
plt.hist(num)
plt.show()
```

```python
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num, color="r", edgecolor="black")
plt.show()
```

```python
# bars ke bich mey width chahiye to
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num, color="r", edgecolor="black",rwidth=0.8)
# yaha hua kya hai ki bhai bars ki pehle width 1 thi ab vo 0.8 ho chuki hai mtlb 0.2 ki gap ban chuki hai
plt.show()
```

```python
# if you want to convert the above table into log table
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num, color="r", edgecolor="black", log=True)
plt.show()
```

```python
# using labael
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num, color="r", edgecolor="black", label="Python")
plt.legend()
plt.show()
```

```python
# bich mey kahi ek line banani ho to axvline parameter use karege
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num, color="r", edgecolor="black", label="Python")
plt.axvline(45, color="cyan", label="Line")
plt.legend()
plt.show()
```

```python
# agar chart mey grid bhi dikhani ho to
plt.xlabel("Python")
plt.ylabel("Number")
plt.title("Yash")
plt.hist(num, color="r", edgecolor="black", label="Python")
plt.axvline(45,color="cyan",label="Line")
plt.grid()
plt.legend()
plt.show()
```

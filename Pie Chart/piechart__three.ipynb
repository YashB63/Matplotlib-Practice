```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ["C","C++","Java","Python"]

plt.pie(x,labels=y)

plt.show()
```

```python
# Abhi dekho chart jo  hai default angle par set hai. Ab agar aap chahte hai ki chart ko ghumaya jaaye to startangle parameter use karo
plt.pie(x,labels=y,startangle=90)

plt.show()
```

```python
# Aur ghumana ho to start angle ko aur change karo
plt.pie(x,labels=y,startangle=180,autopct="%0.2f%%")

plt.show()
```

```python
#font ko badha chota karna ho to textprop parameter use karo and usme fontsize use karo
plt.pie(x,labels=y,startangle=0,autopct="%0.2f%%",textprops={"fontsize":20},radius=2) # overlap kam karne ke liye simply chart ka radius badha diya hai

plt.show()
```

```python
plt.pie(x,labels=y,autopct="%0.2f%%",counterclock=True)

plt.show()
```

```python
# chart ko ghumane ka ek aur tarika
plt.pie(x,labels=y,autopct="%0.2f%%",counterclock=False)

plt.show()
```

```python
# ab lines wagera ko mota karna ho to wedgeprops parameter mey linewidth use karo
ex=[0.0,0.3,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",counterclock=False,shadow=True,wedgeprops={"linewidth":4}) #"width":2 ye bhi use kar skte ho wedge props ke andar

plt.show()
```

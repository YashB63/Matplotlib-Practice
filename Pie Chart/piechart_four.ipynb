```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ["C","C++","Java","Python"]

plt.pie(x,labels=y)

plt.show()
```

```python
# Ab agar aap chaho to edge ke color ko bhi change kar skte ho
# USke liye wedge props ke andar edgecolor ko use karo
ex=[0.0,0.3,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",counterclock=False,shadow=True,wedgeprops={"linewidth":4,"edgecolor":"m"}) 
plt.show()
```

```python
# center ki property change karna ho to 
ex=[0.0,0.3,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",counterclock=False,shadow=True,wedgeprops={"linewidth":4},center=(3,6)) # Center change hone ke baad agar aap dekho to java chart ke andar aa chuka hai
plt.show()
```

```python
# Ab labels ko rotate karna ho to wedgeprop mey rotatelabels use karo
ex=[0.0,0.3,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",counterclock=False,shadow=True,wedgeprops={"linewidth":4},rotatelabels=True) 

plt.show()
```

```python
# title aur legend waala label laga dete hai
ex=[0.0,0.3,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",counterclock=False,shadow=True,wedgeprops={"linewidth":4})
plt.title("Yash")
plt.legend(loc = 5) #loc=5 mtlb location kya rkhni hai vo set kr rhe hai

plt.show()
```

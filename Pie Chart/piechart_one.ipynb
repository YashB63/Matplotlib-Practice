```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ["C","C++","Java","Python"]

plt.pie(x,labels=y)

plt.show()
```

```python
# agar koi tukda bahar chahiye pure pie mey se to what we will do is we will use explode parameter
ex = [0.3,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
# Aur tod fod machani hai to
ex = [0.3,0.0,0.3,0.0]
plt.pie(x,labels=y,explode=ex)
plt.show()
```

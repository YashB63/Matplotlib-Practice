```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ["C","C++","Java","Python"]

plt.pie(x,labels=y)

plt.show()
```

```python
# ab pie chart ka colour change karna ho to
c = ["r","b","g","y"]
plt.pie(x,labels=y,colors=c)

plt.show()
```

```python
# konsa tukda kitne pecentage ka hai vo show karana ho to autopct parameter use karege
c = ["r","b","g","y"]
plt.pie(x,labels=y,colors=c,autopct="%0.1f%%") # is tarike percentage dikhne lagega chart mey 0.1f matlab point ke baad 1  decimal aayega. .2 karoge to 2 zeroess aayege

plt.show()
```

```python
c = ["r","b","g","y"]
plt.pie(x,labels=y,colors=c,autopct="%0.2f%%") # .2 karoge to 2 zeroes aayege

plt.show()
```

```python
# 3D waali feel dena hai to shadow parameter use karo
c = ["r","b","g","y"]
plt.pie(x,labels=y,colors=c,autopct="%0.2f%%",shadow=True) # shadow parameter se pie ke niche shadow aa jaayega
plt.show()
```

```python
c = ["r","b","g","y"]
ex=[0.0,0.3,0.0,0.0]
plt.pie(x,labels=y,colors=c,explode=ex,autopct="%0.2f%%",shadow=True) # shadow parameter se pie ke niche shadow aa jaayega
plt.show()
```

```python
# char ki radius ko badhana ho to radius parameter use karege
c = ["r","b","g","y"]
plt.pie(x,labels=y,colors=c,autopct="%0.2f%%",shadow=True,radius=2) # radius parameter yaha usse karege. is se pie chart ka ssize badh jaayega
plt.show() 
```

```python
# Ab agar aap chahte hai ki bhai chart se labelss ko thoda dur karna hai to use labeldistance parameter
c = ["r","b","g","y"]
plt.pie(x,labels=y,colors=c,autopct="%0.2f%%",shadow=True,labeldistance=1.5) # shadow parameter se pie ke niche shadow aa jaayega
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ["C","C++","Java","Python"]
ex = [0.0,0.3,0.0,0.0]

plt.pie(x,labels=y)

plt.show()
```

```python
plt.pie([1])
plt.show()
```

```python
x = [10,20,30,40]
x1=[40,20,30,10]
y = ["C","C++","Java","Python"]
ex = [0.0,0.3,0.0,0.0]
c = ["r","b","g","y"]

plt.pie(x,labels=y,radius=2)
plt.pie(x1,radius=0.5,colors=c)

plt.show()
```

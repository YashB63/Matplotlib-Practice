```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
```

```python
plt.stem(x,y)
plt.show()
```

```python
# Solid line ko dotted line mey convert karna ho to
```

Linefmt:        character        line style
                
                " '-' "          Solid Line
                " '--' "         Dashed Line
                " '_.' "         Dash-dot Line
                " ':' "          DOtted Line

```python
plt.stem(x,y,linefmt=":") # dotted line
plt.show()
```

```python
plt.stem(x,y,linefmt="-") # ssolid line
plt.show()
```

```python
plt.stem(x,y,linefmt="--") # dashed line
plt.show()
```

```python
plt.stem(x,y,linefmt="-.") # dash-dot line
plt.show()
```

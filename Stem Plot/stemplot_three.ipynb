```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
```

```python
plt.stem(x,y)
plt.show()
```

```python
# abhi jo lines ka color hai usko different color mey dikhana hai to use_line_collection parameter ko false bata kr use karo
plt.stem(x,y,use_line_collection=False)
plt.show()
```

```python
# orientation parameter (horizontal or vertical)
plt.stem(x,y,orientation="horizontal") # default vertical rahega
plt.show()
```

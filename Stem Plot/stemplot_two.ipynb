```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
```

```python
plt.stem(x,y)
plt.show()
```

```python
# Abhi if you want to change the color and symbol of the marker use markerfmt
plt.stem(x,y,markerfmt="r+") # marker red color ka hoga and symbol + ka hoga
plt.show()
```

```python
# abhi chart ki bottom line 0 se start ho rhi hai. agar aap bottom line ke start point ko change karna chahte hai to bottom parameter use karo
plt.stem(x,y,linefmt=":",markerfmt="r+",bottom=9)
plt.show()
```

```python
# Ab agar aap base line ke color ko change karna chahte hai to use kare basefmt parameter
plt.stem(x,y, basefmt="m") # base line red se magenta mey change ho jaayegi
plt.show()
```

```python
# Label Parameter
plt.stem(x,y,label="Python")
plt.legend()
plt.show()
```
